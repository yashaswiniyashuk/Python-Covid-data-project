**Find total countries where total_deaths is greater than 1000000.**

```python
locations = data1[data1['total_deaths'] > 1000000]['location'].unique()
locations
```

```python
locations = data1[data1['total_deaths'] > 1000000]['location'].unique()
#Plotting for each location
for location in locations:
    location_data = data1[(data1['total_deaths'] > 1000000) & (data1['location'] == location)]
    date_counts = location_data.groupby('date')['total_deaths'].sum()  #Aggregate total deaths for each date
    x = range(len(date_counts))    #Use index as x-axis
    y = date_counts.values
    plt.plot(x,y, label= location)
    plt.title("Total Deaths Over Time(Locations with > 1000000 total deaths)")
    plt.xlabel("Number of days since First recoded deaths")
    plt.ylabel("Total Deaths")
    plt.legend()
    plt.show()
```
